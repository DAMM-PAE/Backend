```python
import pandas as pd
```

```python
from enum import Enum

class BarField(Enum):
    NOMBREBAR = 0
    DIRECCION = 1
    NUMEROCALLE = 2
    CIUDAD = 4
    TIPOBAR = 7
```

```python
df = pd.read_excel('../static/data/llistat.xlsx',header=None)
df.head()
```

```python
df.fillna("",inplace=True)
```

```python
df.head()
```

```python
# Primera columna

for i in range(0,len(df)):
    print(df.iloc[i,BarField.NOMBREBAR.value])
    print(df.iloc[i,BarField.DIRECCION.value])
    print(df.iloc[i,BarField.NUMEROCALLE.value])
    print(df.iloc[i,BarField.CIUDAD.value])
    print(df.iloc[i,BarField.TIPOBAR.value])

    break
```
